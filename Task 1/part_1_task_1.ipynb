# Final code to generate output after tokenization and BIO Chunking

```python
def tokenize(text):
    return text.split()
def assign_bio_labels(tokens, entities):
    labels = ['O'] * len(tokens)
    for entity in entities:
        start = entity['start']
        end = entity['end']
        label = entity['labels'][0]  # Assuming each token belongs to a single label
        
        # Find the token positions corresponding to the entity start and end positions
        start_token_idx = None
        end_token_idx = None
        for i, token in enumerate(tokens):
            if start >= len(' '.join(tokens[:i])) and start < len(' '.join(tokens[:i+1])):
                start_token_idx = i
            if end >= len(' '.join(tokens[:i])) and end < len(' '.join(tokens[:i+1])):
                end_token_idx = i + 1  # Include the end token
                
        # Assign BIO labels to tokens
        if start_token_idx is not None and end_token_idx is not None:
            for i in range(start_token_idx, end_token_idx):
                if i == start_token_idx:
                    labels[i] = 'B_' + label
                else:
                    labels[i] = 'I_' + label

    return labels
```

```python
training = {}

for idx, row in train_data.iterrows():
    text = row['data']['text']
    annotations = row['annotations']
    entities = []

    # Extract entities from annotations
    for annotation in annotations:
        for entity in annotation['result']:
            entity_info = {
                'start': entity['value']['start'],
                'end': entity['value']['end'],
                'labels': entity['value']['labels']
            }
            entities.append(entity_info)

    # Tokenize the text
    tokens = tokenize(text)

    # Assign BIO labels
    bio_labels = assign_bio_labels(tokens, entities)

    # Create a dictionary with the required format
    output_dict = {
        'text': text,
        'labels': bio_labels
    }

    # Add the output dictionary to the training dictionary with the case_id as the key
    case_id = row['id']
    training[case_id] = output_dict
```

```python
import json

# Define the file path where you want to save the JSON file
output_file_path = "training_data.json"

# Save the training dictionary to a JSON file
with open(output_file_path, 'w') as json_file:
    json.dump(training, json_file, indent=4)

print("Training data saved to", output_file_path)
```
