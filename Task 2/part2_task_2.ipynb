```python
import json
import re

def generate_bio_labels(text, aspects):
    tokens = re.findall(r"[\w']+|[.,!?;]", text)  # Tokenize while preserving punctuation marks
    bio_labels = ['O'] * len(tokens)  # Initialize all labels as 'O' (Outside)

    for aspect in aspects:
        from_index = aspect['from']
        to_index = aspect['to']
        term = aspect['term']
        polarity = aspect.get('polarity', None)

        # Find the starting index of the term in the tokens list
        start_index = text.find(term[0], from_index)
        start_token_index = len(re.findall(r"[\w']+|[.,!?;]", text[:start_index]))  # Adjusted token index

        # Assign 'B' (Beginning) label to the first token of the term
        bio_labels[start_token_index] = 'B'

        # Assign 'I' (Intermediate) label to the subsequent tokens of the term
        for i in range(start_token_index + 1, start_token_index + len(term)):
            bio_labels[i] = 'I'

    return bio_labels

def generate_bio_dataset(dataset):
    bio_data = {}
    for idx, entry in enumerate(dataset):
        text = entry['raw_words']
        aspects = entry['aspects']
        bio_labels = generate_bio_labels(text, aspects)
        bio_data[idx + 1] = {'text': text, 'labels': bio_labels}
    return bio_data

train_data = json.load(open('Laptop_Review_Train.json'))
val_data = json.load(open('Laptop_Review_Val.json'))
test_data = json.load(open('Laptop_Review_Test.json'))

# Generate BIO dataset for train, validation, and test splits
train_bio_data = generate_bio_dataset(train_data)
val_bio_data = generate_bio_dataset(val_data)
test_bio_data = generate_bio_dataset(test_data)

# Save BIO dataset JSON files for train, validation, and test splits
with open('train_bio_data.json', 'w') as f:
    json.dump(train_bio_data, f, indent=4)

with open('val_bio_data.json', 'w') as f:
    json.dump(val_bio_data, f, indent=4)

with open('test_bio_data.json', 'w') as f:
    json.dump(test_bio_data, f, indent=4)
```
